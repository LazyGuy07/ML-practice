```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,mean_squared_error,mean_absolute_error
df=pd.read_csv('Iris.csv')
df
```

```python
df.info()
```

```python
df=df.drop('Id',axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
for col in df:
    if col!='Species':
        plt.boxplot(df[col])
        plt.title(col)
        plt.show()
```

```python
for col in df:
    if col !='Species':
        q1=np.quantile(df[col],0.25)
        q3=np.quantile(df[col],0.75)
        iqr=q3-q1
        lower=q1-1.5*iqr
        upper=q3+1.5*iqr
        df[col]=df[col].mask((df[col]<lower) | (df[col]>upper),df[col].median())
df.describe()
```

```python
plt.figure(figsize=(12,8))
df.hist()
plt.tight_layout()
plt.show()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
x=df.drop('Species',axis=1)
y=df['Species']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
```

```python
model=KNeighborsClassifier(3)
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
ac = accuracy_score(y_test,pred)
print('Accuracy is :',ac)
```

```python
cm=confusion_matrix(y_test,pred)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,pred))
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
k_values = [3, 4, 5, 6, 7, 8, 9]
for k in k_values:
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(x_train, y_train)
    train_accuracy = knn_classifier.score(x_train, y_train)
    test_accuracy = knn_classifier.score(x_test, y_test)
    print("k =", k, "- Train-set Accuracy:", train_accuracy, "- Test-set Accuracy:", test_accuracy)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.impute import SimpleImputer 


pipeline=make_pipeline(SimpleImputer(strategy='median'),StandardScaler())
x_new=pipeline.fit_transform(x_train)

cv_score=cross_val_score(model,x_new,y_train,cv=7,scoring='accuracy')
cv_score
```

### logistic regression diabetes.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('diabetes.csv')
df
```

```python
df.info()
```

```python
for col in df:
    df[col]=df[col].astype('float64')
df.info()
```

```python
df
```

```python
for col in df:
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
```

```python
for col in df:
    if col !='Outcome':
        q1=np.quantile(df[col],0.25)
        q3=np.quantile(df[col],0.75)
        iqr=q3-q1
        lower=q1-1.5*iqr
        upper=q3+1.5*iqr
        df[col]=df[col].mask((df[col]<lower) | (df[col]>upper),df[col].median())
df.describe()
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
print('accuracy score = ',accuracy_score(y_test,pred))
```

```python
cm = confusion_matrix(y_test,pred)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.tree import DecisionTreeClassifier,plot_tree
import category_encoders as ce
```

```python
df = pd.read_csv('play_tennis.csv')
df
```

```python
df.info()
```

```python
x=df.drop('Play Tennis',axis=1)
y=df['Play Tennis']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)

encoder=ce.OrdinalEncoder(cols=['Outlook','Temperature','Humidity','Wind'])
x_train=encoder.fit_transform(x_train)
x_test=encoder.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
model=DecisionTreeClassifier(criterion='entropy',random_state=42)
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
print('accuracy: ',accuracy_score(y_test,pred))
```

```python
cm =confusion_matrix(y_test,pred)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,pred))
```

```python
plot_tree(model,rounded=True,filled=True)
```
