Insights
1. We can see that the market value decreases in general as we move from centre
forward to back all the way towards the goalkeepers.
2. We can also observe that forwards are more valued, followed by midfielders and
defenders. Goalkeepers have a significantly lower total market value. However,
when we look at the average, defenders are below goalkeepers.
3. Analyzing specific positions within forwards, midfielders, and defenders, we notice
that we have a large number of center-backs, central midfielders, and centerforwards
in the dataset. Due to the higher quantity, these positions also have the
highest sum of market value.
4. However, when we consider the average, left-wingers, second strikers, and centerforwards
have the highest value index.
5. This may be because there is only 1 keeper in the game whereas there are multiple
forwards. Also spectators are more entertained by forwards and tend to follow and
rate them more, thus increasing their overall market value.

### Insights
1. England has the highest number of players in the dataset, followed by France, Spain,
Brazil, and Germany. These same five countries have the highest total market values
due to having more players on the list. This is mainly because of the popularity of
sport in these countries but also the dominance of media in these countries.Average
players are overhyped by the media where there are also many better and
competent players coming out of south american countries.
2. France and Norway have the highest maximum value, as they have the two most
valued players. Egypt has the same maximum and minimum value, along with the
highest average value, because the country has only one player in the list (Salah).
3. Other countries have higher averages because they have few players in the
database, such as South Korea, Norway, Slovakia, and Slovenia.
4. The average cannot be considered a great measurment in this analysis as it can be
deceiving. There can way less number of players from a specific player but at high
value and huge amount of players coming out of other country coming out at varied
ranges.Such case can be seen in Egypt where the only significant player is Mohamed
Salah.

```python
top_2 = df[df['Name'].isin(['Erling Haaland', 'Kylian Mbappé'])]
top_2
```

```python
plt.figure(figsize=(10,5))
plt.title("Goals x Assists")
sns.barplot(x='Name', y='Goals', data=top_2, alpha=0.6, label='Goals')
sns.lineplot(data=top_2['Assists'], alpha=0.2, marker='o',
label='Assists')
plt.grid(True, axis='y', linestyle='--', alpha=1.0)
sns.set_style("dark")
plt.legend()
plt.show()
```

```python
player1 = df.loc[df['Market_value'].idxmax()]
player2 = df.loc[df['Market_value'].nlargest(2).idxmin()]
print('Comparison between the two most valuable players:')
print('Player 1:', player1['Name'])
print('Age:', player1['Age'])
print('Position:', player1['Position'])
print('Country:', player1['Country'])
print('Club:', player1['Club'])
print('Matches:', player1['Matches'])
print('Goals:', player1['Goals'])
print('Assists:', player1['Assists'])
print('Player 2:', player2['Name'])
print('Age:', player2['Age'])
print('Position:', player2['Position'])
print('Country:', player2['Country'])
print('Club:', player2['Club'])
print('Matches:', player2['Matches'])
print('Goals:', player2['Goals'])
print('Assists:', player2['Assists'])
```
